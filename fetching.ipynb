```python
import mysql.connector
import pandas as pd
connection = mysql.connector.connect(
    host = 'localhost',
    user = 'root',
    password = 'Bautroixanh12345',
    database = 'class8'
)
query = """
select 
	"Total sales" as "Category",
	sum(`Total_Revenue`) as "Amount"
from data_main 
where `Year` = 2022 
Union all 
select 
	"Sales deductions" as "Category", 
    sum(`Revenue_deductions`) as "Amount"
from data_main 
where `Year` = 2022
Union all 
select 
	"Net revenue" as "Category", 
    sum(`Net_Revenue`) as "Amount"
from data_main 
where `Year` = 2022
Union all 
select 
	"Total Cost" as "Category", 
    sum(`Cost`) as "Amount"
from data_main
where `Year` = 2022
union all 
select 
	"Gross profit" as "Category", 
    sum(`Gross_Profit`) as "Amount" 
from data_main
where `Year` = 2022 
union all 
select 
	"Selling Expenses" as "Category",
    sum(`Selling_Expenses`) as "Amount" 
from data_main
where `Year` = 2022 
union all 
select 
	"Depreciation and amortization" as "Category", 
    sum(`Depreciation_and_amortization`) as "Amount"
from data_main
where `Year` = 2022 
union all 
select 
	"Salary" as "Category",
    sum(`Salary`) as "Amount"
from data_main
where `Year` = 2022 
union all 
select 
	"Rewards of all kinds" as "Category",
    sum(`Rewards_of_all_kinds`) as "Amount"
from data_main
where `Year` = 2022 
union all 
select 
	"Marketing expenses" as "Category", 
    sum(`Marketing_expense`) as "Amount"
from data_main
where `Year` = 2022 
union all 
select 
	"Tax" as "Category", 
    sum(`Tax(VAT)`) as "Amount"
from data_main
where `Year` = 2022 
union all 
select 
	"General expenses" as "Category", 
    sum(`General_Expenses`) as "Amount"
from data_main
where `Year` = 2022 
union all 
select 
	"Management cost" as "Category",
    sum(`Management_costs(allocation)`) as "Management cost"
from data_main
where `Year` = 2022 
union all 
select 
	"Net profit from business activities" as "Category",
    sum(`Net_profit_from_business_activities`) as "Amount"
from data_main
where `Year` = 2022 
union all 
select 
	"Revenue from financial activities" as "Category",
    sum(`Revenue_from_financial_activities`) as "Amount"
from data_main
where `Year` = 2022 
union all 
select 
	"Financial operating expenses" as "Category",
    sum(`Financial_operating_expenses`) as "Amount"
from data_main
where `Year` = 2022 
union all 
select 
	"Net profits" as "Category", 
    sum(`Net_profits`) as "Amount"
from data_main
where `Year` = 2022 
union all 
select 
	"Other incomes" as "Category", 
    sum(`Other_income`) as "Amount"
from data_main
where `Year` = 2022 
union all 
select 
	"Other costs" as "Category",
    sum(`Other_costs`) as "Amount"
from data_main
where `Year` = 2022 
union all 
select 
	"Other profits" as "Category", 
    sum(`Other_profits`) as "Amount"
from data_main
where `Year` = 2022 
union all 
select
	"Profit before tax" as "Category", 
    sum(`Profits_before_tax`) as "Amount"
from data_main
where `Year` = 2022 
union all 
select 
	"Profit after tax" as "Category",
    sum(`Profit_after_taxes`) as "Amount"
from data_main
where `Year` = 2022 
union all 
select 
	"Final net profits" as "Category", 
    sum(`Net_profit`) as "Amount"
from data_main 
where `Year` = 2022""" 
cursor = connection.cursor()
cursor.execute(query) 
result = cursor.fetchall() 
df = pd.DataFrame(result, columns=['Category', 'Amount'])
df['Amount'] = pd.to_numeric(df['Amount'])
df.to_excel(r'D:\Data Intense\Level 2\Day8\income_statement.xlsx')
```
